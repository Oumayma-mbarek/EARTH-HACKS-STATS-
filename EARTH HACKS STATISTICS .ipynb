```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

the following information are based on the survey of GIZ 

```python
data={'type':['paper','organic','plastic','plastic','plastic','plastic','organic','metal','batteries','organic','organic','plastic',
             'plastic','organic','plastic','organic','batteries','plastic','plastic','plastic','metal','paper','paper','paper','metal',
             'plastic']}
```

```python
data=pd.DataFrame(data)
```

```python
plt.pie(x=data["type"].value_counts()[0:7],
       labels=data["type"].value_counts().index,
         startangle=90,
         autopct='%.1f%%',
        colors=sns.color_palette('Set3',15),
        radius=1.7)
plt.title("Waste Type percentage",fontdict={'fontname': 'Comic Sans MS', 'fontsize':20},y=1.3)
plt.show()
plt.savefig('wastetypepercentage1.png', dpi=300)
```
